```python
import cv2
import matplotlib.pyplot as plt

angio = cv2.imread('ANGIO.png', cv2.IMREAD_GRAYSCALE)
rondelle = cv2.imread('RONDELLE.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(angio, cmap='gray')
plt.show()
hist = cv2.calcHist([angio],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

plt.imshow(rondelle, cmap='gray')
plt.show()
hist = cv2.calcHist([rondelle],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
bajo = 70
alto = 140

thres_angio = cv2.inRange(angio, bajo, alto)
thres_angio[thres_angio == 255] = 128 
thres_rondelle = cv2.inRange(rondelle, bajo, alto)
thres_rondelle[thres_rondelle == 255] = 128  

plt.imshow(thres_angio, cmap='gray')
plt.show()
hist = cv2.calcHist([thres_angio],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

plt.imshow(thres_rondelle, cmap='gray')
plt.show()
hist = cv2.calcHist([thres_rondelle],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

Al hacer el ejercicio con la umbralización doble, nos podemos dar cuenta de que esta permite segmentar una imagen en tres clases de niveles de gris, lo que facilita la separación de áreas con diferentes características. Al efectuar el proceso, puede mejorar la conectividad de objetos cercanos y resaltar los contornos en imágenes borrosas o ruidosas. Sin embargo, presenta limitaciones en la extracción de objetos con formas complejas ya que puede no ser capaz de conectar regiones discontinuas o detectar contornos difusos como los de estas imagenes que se muestran un tanto borrosos. Aunque mejora los contornos al resaltar áreas de contraste, la calidad de los mismos puede verse afectada por el ruido o la falta de bordes bien definidos.
