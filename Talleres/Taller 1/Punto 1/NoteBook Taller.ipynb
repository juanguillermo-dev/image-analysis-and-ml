```python
import cv2
import matplotlib.pyplot as plt

clefG = cv2.imread('CLEF.png', cv2.IMREAD_GRAYSCALE)
muscleG = cv2.imread('MUSCLE.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(clefG, cmap='gray')
plt.show()
plt.imshow(muscleG, cmap='gray')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

hist = cv2.calcHist([clefG],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

hist = cv2.calcHist([muscleG],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

```

Explicación Histogramas:

Para la imagen CLEF, vemos que el pico más alto se encuentra entre los valores de 120 y 140. Esto nos indica que la mayoría de los pixeles que conforman la imagen poseen un nivel de intensidad que se encuentra en ese rango (Estos pixeles son el fondo uniforme de la imagen que es la superficie sobre la cual reposa la llave). Por otro lado, la imagen MUSCLE tiene un pico muy alto y pronunciado justo en un valor un poco más grande que 50, luego una pequeña colina que aciende desde el valor de 100 hasta al rededor de 180 y finalmente dos picos muy bajos pero pronunciados rondando el valor de 240. Esto refleja que en esta imagen de una fibra muscular existe mayor variedad de intensidades para representar las diferentes tonalidades de grises de las diferentes fibras (En la imagen podemos ver que la mayor parte de la imagen esta ocupada por fibras de un gris muy claro, un gris muy oscuro y pocas fibras de una tonalidad intermedia).

El histograma sugiere que existe una gran variedad de escalas de grises a lo largo de la imagen, pero es posible identificar que el pico alto y delgado que se encuentra rondando la intensidad de valor 50 representa las fibras oscuras de baja intenidad y de tonalidad uniforme que podemos apreciar en la imagen. Por otro lado, vemos que las fibras que no son tan oscuras en realidad varían con una intensidad entre 150 y 250, y la extención de la colina que se puede apreciar en el histograma sugiere que son estas fibras de grises claros las que ocupan una mayor superficie de la imagen.
