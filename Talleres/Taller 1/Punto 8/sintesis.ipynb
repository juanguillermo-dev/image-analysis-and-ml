```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


muscle = cv2.imread("MUSCLE.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(muscle, cmap='gray')
plt.show()
```

```python
hist_muscle = cv2.calcHist([muscle],[0],None,[256],[0,256])
plt.plot(hist_muscle)
plt.show()
```

```python
intensidades = 90

plt.imshow(muscle, cmap='gray')
plt.show()


i_arr = [[255 if (muscle[j][i] <= intensidades) else muscle[j][i]  for i in range(256)] for j in range(256)]



plt.imshow(i_arr, cmap='gray', vmin=0, vmax=255)
plt.show()

```

```python
# Aquí utilizamos solo operandos de la imagen original

i = [[200 for _ in range(256)] for _ in range(256)]
j = [[255 for _ in range(256)] for _ in range(256)]
n = [[1 for _ in range(256)] for _ in range(256)]


i_arr = np.array(i, dtype=np.uint8).reshape(256, 256)
j_arr = np.array(j, dtype=np.uint8).reshape(256, 256)
n_arr = np.array(n, dtype=np.uint8).reshape(256, 256)

modified = cv2.bitwise_or(muscle, i_arr)
modified2 = cv2.subtract(j_arr, muscle)
modified3 = cv2.subtract(modified2,muscle)
modified4 = cv2.subtract(modified3,muscle)
modified6 = cv2.add(modified3,muscle)
modified7 = cv2.bitwise_and(modified2,modified6)
modified_l1 = cv2.add(muscle,modified7) 

plt.imshow(modified_l1, cmap='gray', vmin=0, vmax=255)
plt.show()

```
