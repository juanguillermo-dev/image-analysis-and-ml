```python
import cv2
import matplotlib.pyplot as plt
import numpy as np



bureau = cv2.imread("BUREAU.png", cv2.IMREAD_GRAYSCALE)

```

```python
plt.imshow(bureau, cmap='gray')
plt.show()
```

```python
def bit_plane_slicing(imagen, k):
    plane = np.full((imagen.shape[0], imagen.shape[1]), 2 ** k, np.uint8)
    return cv2.bitwise_and(plane, imagen)
```

```python
plano_7 = bit_plane_slicing(bureau, 7)
plt.imshow(plano_7, cmap='gray')
plt.title('Bit Plane 7')
plt.show()

plano_6 = bit_plane_slicing(bureau, 6)
plt.imshow(plano_6, cmap='gray')
plt.title('Bit Plane 6')
plt.show()

plano_5 = bit_plane_slicing(bureau, 5)
plt.imshow(plano_5, cmap='gray')
plt.title('Bit Plane 5')
plt.show()

plano_4 = bit_plane_slicing(bureau, 4)
plt.imshow(plano_4, cmap='gray')
plt.title('Bit Plane 4')
plt.show()

plano_3 = bit_plane_slicing(bureau, 3)
plt.imshow(plano_3, cmap='gray')
plt.title('Bit Plane 3')
plt.show()

plano_2 = bit_plane_slicing(bureau, 2)
plt.imshow(plano_2, cmap='gray')
plt.title('Bit Plane 2')
plt.show()

plano_1 = bit_plane_slicing(bureau, 1)
plt.imshow(plano_1, cmap='gray')
plt.title('Bit Plane 1')
plt.show()

plano_0 = bit_plane_slicing(bureau, 0)
plt.imshow(plano_0, cmap='gray')
plt.title('Bit Plane 0')
plt.show()
```

Se identificaron desde el plano más significativo (plano 7) y el menos significativo (plano 0). De estos planos se puede concluir que en la medida que disminuyen los planos aumenta el ruido, es decir, las imagenes pierden forma y aparece una grann cantidad de puntos negros con blanco. Por otro lado, en la medida que aumenta el plano se pueden apreciar colores más intensos, llegando a un punto de que se puede ver una sombra fluida de la imagen. Por último, se puede apreciar que es en los planos intermedios en los que se evidencia mayor detalle, como en el plano 5 que se puede entender bien el contenido de la imagen. Esto se puede deber a la escala de intensidad utilizada.

```python
def image_quantization(imagen, k):
    plane = np.full((imagen.shape[0], imagen.shape[1]), 255-2 ** k, np.uint8)
    return cv2.bitwise_and(plane, imagen)
```

```python
cuant5 = image_quantization(bureau, 5)
plt.imshow(cuant5, cmap='gray')
plt.title('Quantization k=5')
plt.show()

cuant3 = image_quantization(bureau, 3)
plt.imshow(cuant3, cmap='gray')
plt.title('Quantization k=3')
plt.show()

cuant1 = image_quantization(bureau, 1)
plt.imshow(cuant1, cmap='gray')
plt.title('Quantization k=1')
plt.show()
```

Se puede apreciar que cuando la cuantificación tiene k=1 se empieza a ver difuminada la imagen. Luego, con k = 3 se empiezan a ver contornos falsos, en especial en zonas que antes tenian un color uniforme como lo puede ser el escritorio. Finalmente, con k = 5 se aprecian aun más contornos falsos e incluso se puede ver pixelada la foto. 

Esto se explica ya que cada vez existen menos colores para interpretar la imagen, por lo que se empiezan a unificar las zonas de intensidad por lo que empiezan a surgir contornos falsos en zonas que antes contaban con más colores para expresarse.
