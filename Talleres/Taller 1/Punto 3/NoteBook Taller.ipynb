```python
import cv2
import matplotlib.pyplot as plt

regleG = cv2.imread('REGLE.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(regleG, cmap='gray')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

hist = cv2.calcHist([regleG],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

# ¿Cómo es la dinámica de la imagen?

Podemos destacar varios detalles con respecto al histograma:
1. La imagen no posee pixeles con una intensidad superior a 120.
2. El mayor grupo de pixeles de igual tonalidad tiene una intensidad de entre 100 y 110.
3. En el rango entre las intensidades de 60 a 110, la cantidad de pixeles aumenta drasticamente segun aumenta la intensidad

```python
negativoG = cv2.bitwise_not(regleG)
plt.imshow(negativoG, cmap='gray')
plt.show()
```

```python
histN = cv2.calcHist([negativoG],[0],None,[256],[0,256])
plt.plot(histN)
plt.show()
```

# ¿Cómo es este histograma con respecto al de la imagen original?

Al realizar el negativo de la imagen y obtener su histograma, podemos ver que ha sucedido un efecto de espejo sobre la gráfica. El negativo vuelve los negros blancos y viseversa y el mismo efecto surge con la escala de grises. Es como si dividieramos los niveles de intensidad a la mitad y los pixeles pasaran de tomar su intensidad de un lado de la linea a esa misma intensidad pero del otro lado, haciendo que los grises oscuros se vuelvan claro y los claros se vuelvan oscuros.

```python
import numpy as np

minG = np.min(regleG)
maxG = np.max(regleG)
expandidaG = ((regleG - minG) / (maxG - minG) * 255).astype(np.uint8)

plt.imshow(negativoG, cmap='gray')
plt.show()
plt.imshow(expandidaG, cmap='gray')
plt.show()
```

```python
histE = cv2.calcHist([expandidaG],[0],None,[256],[0,256])
plt.plot(histE)
plt.show()
```
