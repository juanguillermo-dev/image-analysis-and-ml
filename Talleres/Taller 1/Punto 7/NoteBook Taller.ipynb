```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

rampa = np.linspace(0, 255, 256)
imagen = np.tile(rampa, (256, 1))

plt.imshow(imagen, cmap='gray')
plt.show()
```

```python
aquitainG = cv2.imread('AQUITAIN.png', cv2.IMREAD_GRAYSCALE)
scannerG = cv2.imread('SCANNER.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(aquitainG, cmap='gray')
plt.show()
plt.imshow(scannerG, cmap='gray')
plt.show()
```

```python
import matplotlib.pyplot as plt

im1 = plt.imshow(imagen, cmap='viridis')
plt.colorbar(im1)
plt.show()
plt.hist(imagen.ravel(), bins=256)
plt.show()

im2 = plt.imshow(aquitainG, cmap='viridis')
plt.colorbar(im2)
plt.show()
plt.hist(aquitainG.ravel(), bins=256)
plt.show()

im3 = plt.imshow(scannerG, cmap='viridis')
plt.colorbar(im3)
plt.show()
plt.hist(scannerG.ravel(), bins=256)
plt.show()
```

# Observasiones:

Lo que podemos ver de las imagenes AQUITAIN y SCANNER basandonos en el histograma de la rampa a color, es que la primera tiene un contraste más bajo mientras la segunda tiene un contraste más alto. Al ver el histograma de AQUITAIN nos damos cuentas que la imagen a color cuenta con unas intensidades distribuidas entre 10 y 35 y varían entre estas paulatinamente. Por otro lado, el histograma de SCANNER muestra que la imagen es de alto contraste y la intensidad varía entre 0 y 80, con un salto abrupto entre 15 y 25.

```python
scannerFloat = scannerG.astype(float)
min = scannerFloat.min()
max = scannerFloat.max()
expandida = ((scannerFloat - min) * (255.0 / (max - min))).astype(np.uint8)

im4 = plt.imshow(expandida, cmap='grey')
plt.colorbar(im4)
plt.show()
plt.hist(expandida.ravel(), bins=256)
plt.show()

im5 = plt.imshow(expandida, cmap='viridis')
plt.colorbar(im5)
plt.show()
plt.hist(expandida.ravel(), bins=256)
plt.show()
```

# Observasiones:

Las diferencias con la imagen original son casi nulas, sin embarbargo, es posible ver que se acentuan un poco los contrastes en el centro de la imagen. No se muy bien que es lo que muestra la imagen (parece ser una célula o una bacteria), pero en el centro del objeto principal hay una región más oscura que guarda algo de detalles en su interior. Estos detalles tienen un leve mejor visibilidad después de aplicar la expansión.
